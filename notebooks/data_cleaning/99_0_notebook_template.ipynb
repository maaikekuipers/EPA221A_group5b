```python
# parameters
# change to raw data file
INPUT_DATA_PATH = "data/raw/.csv"
INPUT_SHAPEFILE_PATH = "data/gemeente_shapefiles/gemeente_shapefile.shp"

# set true or false
SAVE_OUTPUT = False
# change to filename
OUTPUT_DATA_PATH = "data/processed/filename.csv"
```

```python
import numpy as np
import pandas as pd
import geopandas as gpd

from pathlib import Path
```

```python
gdf = gdf.to_crs("EPSG:4326")
```
